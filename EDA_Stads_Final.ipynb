# function to convert datestring with time into time and date

```python
def convert_2_datetime(date_string):
    from datetime import datetime
    
    # Given string
    #date_string = "29-03-2023 12:00:13"
    try:
    # Convert string to datetime object
        date_time_obj = datetime.strptime(date_string, '%d-%m-%Y %H:%M:%S')
    except:
        date_time_obj = datetime.strptime(date_string, '%d-%m-%Y %H:%M')
    return date_time_obj
```

```python
st_date=[]
st_time=[]
st_year = []
for i in start_time:
    st_date.append(convert_2_datetime(i).date())
    st_time.append(convert_2_datetime(i).time())
    st_year.append(convert_2_datetime(i).date().year)
df["Start Time_date"] = st_date
df["Start Time_time"] = st_time
df["Start Time_year"] = st_year
```

```python
def total_duration_in_mins_for_TV_show_per_year(tv_show_Df, years):
    tot_dur = []
    for i in range(len(years)):
        df=tv_show_Df.loc[tv_show_Df['Start Time_year']==years[i], 'duration_datetime']
        t=int(total_mins_duration(df))
        tot_dur.append(t)
    return tot_dur
```

```python
import matplotlib.pyplot as plt
from datetime import datetime

# Given data
years = [2018, 2019, 2020, 2021, 2022, 2023]
start_times = ['00:13:22', '00:00:23', '00:00:00', '00:00:01', '07:59:59', '00:16:15']
end_times = ['23:33:55', '23:59:53', '23:59:41', '23:59:12', '22:44:48', '23:55:43']
start_time_titles = ['Shadowhunters','Prison Break','One-Punch Man','Sex Education','The Blacklist','Anatomie eines Skandals']
end_times_titles = ['Tote Mdchen lgen nicht','Prison Break','One-Punch Man','Naruto Shippuden','The Witcher','Staffel 2 (Trailer): Ginny & Georgia']
# Convert start and end times to datetime objects
start_times_dt = [datetime.strptime(time, '%H:%M:%S') for time in start_times]
end_times_dt = [datetime.strptime(time, '%H:%M:%S') for time in end_times]

# Extract hours, minutes, and seconds
start_times_str = [time.strftime('%H:%M:%S') for time in start_times_dt]
end_times_str = [time.strftime('%H:%M:%S') for time in end_times_dt]

# Plotting the graph
plt.figure(figsize=(15, 7))
plt.plot(years, [t.hour + t.minute / 60 + t.second / 3600 for t in start_times_dt], marker='o', label='Earliest Time')
plt.plot(years, [t.hour + t.minute / 60 + t.second / 3600 for t in end_times_dt], marker='o', label='Latest Time')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Time of Day')
plt.title('Earliest and Latest Times for Each Year')

# Set y-axis ticks with 30-minute intervals
plt.yticks([i/2 for i in range(48)], ['{:02d}:{:02d}:{:02d}'.format(i//2, (i%2)*30, 0) for i in range(48)])

# Add text over each point
for year, start_time, end_time, st_title, et_title in zip(years, start_times_str, end_times_str, start_time_titles, end_times_titles):
    plt.text(year, float(start_time[:2]) + float(start_time[3:5]) / 60 + float(start_time[6:]) / 3600, f"{st_title}: {start_time}", ha='center', va='bottom')
    plt.text(year, float(end_time[:2]) + float(end_time[3:5]) / 60 + float(end_time[6:]) / 3600, f"{et_title}: {end_time}", ha='center', va='bottom')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
